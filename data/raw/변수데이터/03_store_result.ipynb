```python
import pandas as pd

df_code = pd.read_csv("data/01_store_code.csv")
df_sale = pd.read_csv("data/02_store_sale.csv")
```

```python
# df_sale 기준으로 df_code 병합
df_result = pd.merge(
    df_sale,
    df_code[['region_code', 'district_code', 'district', 'longitude', 'latitude']],
    on='region_code',
    how='left'
)
```

```python
# df_code에 매칭되지 않아 district_code가 NaN인 행 제거
df_result = df_result.dropna(subset=['district_code']).reset_index(drop=True)
```

```python
# 두 데이터프레임 조인 후 불필요해진 컬럼제거
df_result = df_result.drop(columns=['district_code'])
```

```python
# 컬럼값 영문으로 일괄 변경
df_result['district'] = df_result['district'].replace({
    '은평구': 'eunpyeong',
    '구로구': 'guro'
})
```

```python
# 0이 있는 행 확인
zero_rows = (df_result == 0).any(axis=1)  # 행 기준으로 하나라도 0이면 True

# 0이 있는 행 제거
df_result = df_result[~zero_rows]
```

```python
output_path = r'data/03_store_result.xlsx'
df_result.to_excel(output_path, index=False)
```
